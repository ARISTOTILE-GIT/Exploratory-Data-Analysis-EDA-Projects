# 12.Extract the genres (note: some movies may have multiple genres, split them).

```python
from collections import Counter
import ast
```

```python
import ast
from collections import Counter
import pandas as pd

def extract_genre_names(genre_str):
    try:
        genres = ast.literal_eval(genre_str)
        return [genre['name'] for genre in genres]
    except (SyntaxError, ValueError):
        return ['unknown']

df['genres_list'] = df['genres'].apply(extract_genre_names)

all_genres = [genre for sublist in df['genres_list'] for genre in sublist]

genre_counts = Counter(all_genres)

genre_counts_df = pd.DataFrame(genre_counts.items(), columns=['Genre', 'Count'])
```

# 13.Count the number of movies in each genre.

```python
genre_counts_df
```

# 14.Plot a bar chart of the movie counts per genre.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(genre_counts_df['Genre'], genre_counts_df['Count'])
plt.title("GENRES OVER COUNTS")
plt.xticks(rotation=90)

for index, value in enumerate(genre_counts_df['Count']):
    plt.text(index, value + 20, str(value), ha="center", va="bottom", fontsize=9, fontweight="bold")

plt.tight_layout()
plt.show()
```
